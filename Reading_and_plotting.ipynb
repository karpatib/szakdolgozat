```python
def plot3(X,Y,Z,mark="o",col="r", ):
  from matplotlib import pyplot
  import pylab
  from mpl_toolkits.mplot3d import Axes3D
  pylab.ion()
  fig = pylab.figure()
  ax = Axes3D(fig)
  ax.scatter(X, Y, Z,marker=mark,color=col,)
```

```python
plot3(z,x,y,mark=".",col="r")
plt.grid()
```

```python
plot3(z1,x1,y1,mark=".",col="b")
```

```python
plot3(sz_x,sz_y,sz_z,mark=".",col="y")
```

```python
plot3(sz_x1,sz_y1,sz_z1,mark=".",col="g")
```
